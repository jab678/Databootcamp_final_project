```python
df1=pd.read_csv('C:\\Users\\Joshua Bienstock\\Desktop\\DataUsa_Main.csv')
df2=pd.read_csv('C:\\Users\\Joshua Bienstock\\Desktop\\DataUsa_Sins.csv')
df3=pd.read_csv('C:\\Users\\Joshua Bienstock\\Desktop\\DataUsa_ChildStats.csv')
```

```python
#Pulls only County level data 
df1_county = df1[df1['geo_sumlevel']=='county']
```

```python
#Sets all index to Geo_id
df1_county=df1_county.set_index('geo_id')
df2=df2.set_index('geo_id')
df3=df3.set_index('geo_id')
```

```python
#combines DFs
final=pd.merge(df1_county, df2, left_index=True, right_index=True)
final=pd.merge(final, df3, left_index=True, right_index=True)
```

```python
##Deletes 2015, 2016 columns
columnlist=final.columns
columnlist=columnlist.tolist()
removelist=[]
for i in columnlist:
    if('2015' in i):
        removelist.append(i)
    elif('2016' in i):
        removelist.append(i)
```

```python
for i in removelist:
    final = final.drop(i,1)
```
